```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# Load environment variables in a file called .env

load_dotenv(override=True)
API_KEY = os.getenv('OPENAI_API_KEY')

# Check the key
if (not API_KEY) or (not isinstance(API_KEY, str)):
    print("No API key was found - please head over to the troubleshooting notebook in this folder to identify & fix!")
elif (not API_KEY.startswith("sk-proj-")):
    print("An API key was found, but it doesn't start sk-proj-; please check you're using the right key - see troubleshooting notebook")
elif (API_KEY.strip() != API_KEY):
    print("An API key was found, but it looks like it might have space or tab characters at the start or end - please remove them - see troubleshooting notebook")
else:
    print("API key found and looks good so far!")

openai = OpenAI(api_key=API_KEY)
```

## Experimenting with temperature and max_output_tokens

- Intelligence that Innovates, Tomorrow that Transforms.  
- Empowering Tomorrow, Intelligently Today.


```python
prompt = "Write a creative tagline for an AI startup."
```

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    max_output_tokens=30
)

display(Markdown(response.output_text))
```

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    temperature=0.2,
    max_output_tokens=30
)

display(Markdown(response.output_text))
```

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    temperature=0.9,
    max_output_tokens=30
)

display(Markdown(response.output_text))
```

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    top_p=0.6,
    temperature=1.0,
    max_output_tokens=30
)

display(Markdown(response.output_text))
```

```python
top_ps = [0.2, 0.5, 0.8, 0.95]

for p in top_ps:
    
    response = openai.responses.create(
        model="gpt-4.1-mini",
        input=prompt,
        top_p=p,
        temperature=1.0,
        max_output_tokens=30
    )
    
    display(Markdown(f"\nTop_p = {p}"))
    display(Markdown(response.output_text))
```

## Zero-Shot Prompting

```python
prompt = """
Classify the sentiment of the following review as Positive, Negative, or Neutral.

Review: The battery life is amazing but the screen scratches easily.

Sentiment:
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    temperature=0.2,
    max_output_tokens=20
)

display(Markdown(response.output_text))
```

```python
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content": """
Classify the sentiment of the following review as Positive, Negative, or Neutral.

Review: The battery life is amazing but the screen scratches easily.

Sentiment:
"""
        }
    ],
    temperature=0.2,
    max_tokens=20
)
display(Markdown(response.choices[0].message.content))
```

```python
prompt = """
Translate the following sentence to French:

Hello, how are you today?
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt
)

display(Markdown(response.output_text))
```

## Few-Shot Prompting

```python
prompt = """
Classify the sentiment of the following reviews.

Text: I love the camera quality.
Sentiment: Positive

Text: The phone overheats frequently.
Sentiment: Negative

Text: The delivery was on time.
Sentiment: Neutral

Text: The interface is confusing but usable.
Sentiment:
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    temperature=0.2,
    max_output_tokens=20
)

display(Markdown(response.output_text))
```

```python
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {
            "role": "user",
            "content": """
Classify the sentiment of the following reviews.

Text: I love the camera quality.
Sentiment: Positive

Text: The phone overheats frequently.
Sentiment: Negative

Text: The delivery was on time.
Sentiment: Neutral

Text: The interface is confusing but usable.
Sentiment:
"""
        }
    ],
    temperature=0.2,
    max_tokens=20
)

display(Markdown(response.choices[0].message.content))
```

```python
prompt = """
Generate a Python function based on the description.

Description: Adds two numbers
Function:
def add(a, b):
    return a + b

Description: Multiplies two numbers
Function:
def multiply(a, b):
    return a * b

Description: Calculates the area of a rectangle
Function:
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    max_output_tokens=100
)

display(Markdown(response.output_text))
```

```python
prompt = """
Translate English to Spanish.

English: Hello
Spanish: Hola

English: Good morning
Spanish: Buenos días

English: Thank you
Spanish:
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt
)

display(Markdown(response.output_text))
```
