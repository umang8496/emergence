## Basic Prompting

```python
message = """
Summarize the following text in two sentences:

Large Language Models are AI systems trained on large text datasets.
They can perform tasks such as summarization, translation, classification, and code generation.
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=message,
    temperature=0.3
)

display(Markdown(response.output_text))
```

## The Effect of Temperature

The **temperature** parameter controls the randomness or creativity of the output.  
A **lower temperature** makes the output more focused and deterministic, while a **higher temperature** leads to more diverse and sometimes unexpected results.  

**Task**: Generate ideas for a fantasy story plot.

```python
message = """Brainstorm three plot ideas for a fantasy story involving a lost map and a hidden city."""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=message,
    temperature=0.9
)

display(Markdown(response.output_text))
```

```python
message = """Brainstorm three plot ideas for a fantasy story involving a lost map and a hidden city."""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=message,
    temperature=0.1
)

display(Markdown(response.output_text))
```

## Managing Output Length with `max_tokens`

```python
prompt = "Explain the process of photosynthesis in simple terms."

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    temperature=0.5,
    max_output_tokens=100
)

display(Markdown(response.output_text))
```

## Custom Responses

### Basic Request Structure

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input="Explain the concept of recursion in programming."
)

display(Markdown(response.output_text))
```

### Using System Instructions (Behavior Control)

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    instructions="You are a senior software engineer who explains concepts clearly.",
    input="Explain recursion in programming."
)

display(Markdown(response.output_text))
```

### Multi-Message Conversations

```python
response = openai.responses.create(
    model="gpt-4.1-mini",
    input=[
        {"role": "system", "content": "You are a helpful tutor."},
        {"role": "user", "content": "Explain what a neural network is."}
    ]
)

display(Markdown(response.output_text))
```

### Structured Outputs

```python
message = """
Extract the following fields from the text.

Text: John works at Google.

Return JSON with fields:
name
company
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=message
)

display(Markdown(response.output_text))
```

### Penalties to Reduce Repetition

- frequency_penalty=0.5 [Reduces repeating tokens]
- presence_penalty=0.5 [Encourages introducing new topics]

```python
response = openai.chat.completions.create(
    model="gpt-4.1-mini",
    messages=[
        {"role": "user", "content": "Explain the Middle-East crisis in simple terms."}
    ],
    frequency_penalty=0.5,
    presence_penalty=0.5,
    max_tokens=200
)

display(Markdown(response.choices[0].message.content))
```

### Top-k Sampling

Top-k sampling restricts the model to selecting from the **K** most probable tokens.  

How it works:

- The model calculates probabilities for all tokens.
- Tokens are sorted by probability.
- Only the top K tokens are kept.
- All other tokens are discarded.
- The next token is sampled from these K tokens.

Advantages:

- Easy to understand
- Strictly limits candidate tokens
- Prevents extremely unlikely tokens

Limitations:

- Fixed cutoff
- May remove good tokens if k is too small
- May still allow bad tokens if distribution is flat

#### OpenAI does not support top_k
#### Other framework (Hugging Face) do support top_k

### Top-p (Nucleus Sampling)

Top-p sampling restricts tokens based on cumulative probability instead of a fixed number.  

Instead of choosing `K` tokens, it chooses the smallest set of tokens whose total `probability ≥ p`.  

How it works:

- Sort tokens by probability.
- Add tokens until cumulative probability ≥ p.
- Only those tokens form the candidate set.
- Sample from that set.

Advantages:

- Adaptive
- Keeps only the meaningful probability mass
- Prevents sampling from rare "tail tokens"

This is why `top_p` is generally preferred over `top_k` in modern systems.

```python
prompt = """
Suggest three innovative uses for graphene.
"""

response = openai.responses.create(
    model="gpt-4.1-mini",
    input=prompt,
    top_p=0.85,
    temperature=1.0,
    max_output_tokens=100
)

display(Markdown(response.output_text))
```

```python
for p in [0.5, 0.8, 0.95]:
    
    response = openai.responses.create(
        model="gpt-4.1-mini",
        input="Write a creative title for a science fiction novel.",
        top_p=p,
        temperature=1.0,
        max_output_tokens=50
    )
    
    display(Markdown(f"\nTop_p = {p}"))
    display(Markdown(response.output_text))
```

#### Top-k vs Top-p

| Feature            | Top-k        | Top-p            |
| ------------------ | ------------ | ---------------- |
| Token selection    | Fixed number | Probability mass |
| Adaptive           | No.          | Yes.             |
| Stability          | High         | Moderate         |
| Creativity control | Limited      | Better           |
| Industry usage     | Less common  | More common      |
